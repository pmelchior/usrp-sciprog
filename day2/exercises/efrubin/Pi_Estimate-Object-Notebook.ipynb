```python
%matplotlib inline
import pi_estimate.pi_calc as pie
```

First we just want to estimate pi using 100, 1000, 1e4, and 1e6 points.

```python
points = [100, 1000, 1e4, 1e6]
for p in points:
    print pie.PiGuess(p).pi
```

If we want to visualize each pair, we can use the plot_pairs method of the PiGuess class.

```python
pie.PiGuess(1000).plot_pairs()
```

We then want to determine the calculation time.  The PiGuess class includes a time method, so doing so is simple.

```python
for p in points:
    print pie.PiGuess(p).time
```

We could have obtained the points and times with only one calculation of each pi estimate.  Indeed, that is what's done in the PiGuess_N class.  Now we want to make a plot of how the calculation time depends on the number of points.  We probably want more than four.  We'll use the plot_time method of the PiGuess_N class.

```python
points = [10**(.1*i) for i in range(0,60)]
many_n = pie.PiGuess_N(points)
many_n.plot_time()
```

We'd also like to plot the precision of the estimates with respect to the number of points.  We'll use the plot_precision method of the PiGuess_N class.

```python
many_n.plot_precision()
```

Now we want to estimate pi many times using a single number of points. This is what the PiGuess_Single class does.

```python
many_time = pie.PiGuess_Single(10000,1000)

print many_time.mean
print many_time.std
```

Then we can make a histogram of the data using the histogram method.

```python
many_time.histogram()
```
