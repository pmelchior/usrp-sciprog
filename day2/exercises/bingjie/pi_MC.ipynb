```python
pt_list = [100, 1000, 10000, 1000000]
for pt in pt_list:
    a = pi_estimate.MC(N=pt, time=True)
    print 'Pi is estimated to be', a.pi()[0], ', using', pt, 'points.', '(Time used:', a.pi()[1], 's.)'
```

```python
time_list = []
pi_list = []

pt_list = [100, 1000, 10000, 1000000]
for pt in pt_list:
    a = pi_estimate.MC(N=pt, time=True)
    pi_list.append(a.pi()[0])
    time_list.append(a.pi()[1])
```

```python
plt.plot(pt_list, time_list, 'o')
plt.xscale('log')
plt.yscale('log')
plt.xlim(10, 1e7)
plt.xlabel('Number of points')
plt.ylabel('Time')
```

Make a figure that displays the "darts."

Run this calculation many times for a single N (number of darts), and plot a histogram of the results.

```python
pi_histlst = []

for i in xrange(1000):
    a = pi_estimate.MC(N=100, time=False)
    pi_histlst.append(a.pi())
```

```python
plt.hist(pi_histlst)
```

Calculate the mean and the standard deviation of these estimates.

```python
np.mean(pi_histlst)
```

```python
np.std(pi_histlst)
```

Make a plot that illustrates the precision of the pi estimate as a function of number of random points used.

```python
#pi_list = []
err_list = []

pt_list = np.linspace(100, 100000, 21)
for pt in pt_list:
    a = pi_estimate.MC(N=pt, time=False)
    #pi_list.append(a.pi())
    err_list.append(abs(np.pi - a.pi()))
```

```python
plt.plot(pt_list, err_list, 'o')
plt.xlabel('Number of points')
plt.ylabel('Errors (abs value)')
```
