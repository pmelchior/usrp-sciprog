```python
# dict = {}
wordlst = []
countlst = []
for i in np.unique(larger_than_3):
    word =  i.strip('"')
    count =  larger_than_3.count(word)
#     print "{}:{}".format(word,count)
    wordlst.append(word)
    countlst.append(count)
```
