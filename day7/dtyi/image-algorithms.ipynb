## Reading Images in Python

The image above is a PNG, and while that's a great format to look at (JPEG isn't bad either), it's not what we'd use to run algorithms on, because typically we destroy some information when creating files in that format (either because there's lossy compression or a smaller number of bits per pixel or both).  The FITS version of the image above is a bit too big to fit comfortably in the git repository, so we'll focus on a smaller image, with just a few objects.  Here's how we load it:

```python
import astropy.io.fits
import numpy

fits = astropy.io.fits.open("../data/hsc-image-small.fits")
image = fits[0].data
```

Unlike with tables, here we *can* use the first HDU (HDU 0).

Here's what the smaller image looks like:

```python
from matplotlib import pyplot
%matplotlib inline
pyplot.imshow(image, origin='lower', interpolation='nearest', cmap=pyplot.cm.gray, vmin=-0.1, vmax=1.0)
```

It's worth taking a closer look at some of the options I passed to `imshow` to get a good view of the image, as you'll probably use some of them frequently if you're looking at images:
 - `origin='lower'`: `imshow` usually puts y=0 at the top (the convention for computer scientists) rather than at the bottom (the convention for physical sciences).
 - `interpolation='nearest`: by default, `imshow` will try to be clever and smooth the image for you, but in astronomy we actually want to see the pixels.  "nearest" means "nearest neighbor interpolation".
 - `cmap`: sets the colormap used to map a range of numbers to a range of colors (lots to choose from!)
 - `vmin`, `vmax`: if we just let `imshow` use the minimum and maximum of the array, we won't be able to see any of the fainter stuff at all, because it has a huge dynamic range (and that's why we've saturated the center of the bright galaxy).

## Background Modeling

Both of the images we've looked at here have already had the sky background subtracted from them (in fact, they're actually combinations of many separate exposures, each of which was background-subtracted before they were combined.

We can see this if we look at the variance image:

```python
pyplot.imshow(variance, origin='lower', interpolation='nearest', cmap=pyplot.cm.gray)
```

We can clearly see the boundaries of some of the inputs, including how some parts of the image are deeper than other parts.

Background modeling is fairly straightforward: we mask out the regions where we've detected objects, then compute the mean of what's left over.  We then repeat that over different regions (because the background isn't completely constant), and either interpolate between them or fit a smooth function across them.  It's only tricky because we need to detect the sources before we estimate the background, and we need to subtract the background before we can detect sources.

## Measurement

What properties of astronomical sources might we want to measure?

### Moments

The zeroth moment of an image is just its sum:
$$m_0 = \sum_{x,y} I(x,y)$$

The first moments are:
$$m_x = \frac{1}{m_0}\sum_{x,y} x I(x,y)$$
$$m_y = \frac{1}{m_0}\sum_{x,y} y I(x,y)$$

And the second moments are:
$$m_{xx} = \frac{1}{m_0}\sum_{x,y} (x-m_x)^2 I(x,y)$$
$$m_{yy} = \frac{1}{m_0}\sum_{x,y} (y-m_y)^2 I(x,y)$$
$$m_{xy} = \frac{1}{m_0}\sum_{x,y} (x-m_x)(y-m_y) I(x,y)$$

If we could make a *non-binary* decision about which pixels to include, the moments would look like this (assuming $\sum_{x,y}w(x,y)=1$):

$$m_0 = \sum_{x,y} w(x,y)\;I(x,y)$$

$$m_x = \frac{1}{m_0}\sum_{x,y} x\;w(x,y)\;I(x,y)$$
$$m_y = \frac{1}{m_0}\sum_{x,y} y\;w(x,y)\;I(x,y)$$

$$m_{xx} = \frac{1}{m_0}\sum_{x,y} (x-m_x)^2\;w(x,y)\;I(x,y)$$
$$m_{yy} = \frac{1}{m_0}\sum_{x,y} (y-m_y)^2\;w(x,y)\;I(x,y)$$
$$m_{xy} = \frac{1}{m_0}\sum_{x,y} (x-m_x)(y-m_y)\;w(x,y)\;I(x,y)$$

### Optimal Weights

The variance on the zeroth moment is:

$$\sigma^2_0 = \sum_{x,y} w(x,y)\;V(x,y)$$

where $V(x,y)$ is the variance image.

(this is just standard [propagation of uncertainty](https://en.wikipedia.org/wiki/Propagation_of_uncertainty) - just accept it for now - we'll get into more detail on noise in images tomorrow)

That means the *signal-to-noise ratio* on the zeroth moment is:

$$\mathrm{SNR} = \frac{m_0}{\sigma_0} = \frac{\sum_{x,y}w(x,y)\;I'(x,y)}{\sum_{x,y}w(x,y)\;V(x,y)}$$.

Where $I'(x,y)$ is the *ideal* image - the image without noise (which we obviously don't actually have).

Let's imagine for a moment that the variance image is constant - that's common for faint objects, because most of the noise from the sky background.  That means:

$$\mathrm{SNR} \propto \sum_{x,y}w(x,y)\;I'(x,y)$$.

If we recall that

$$\sum_{x,y}w(x,y) = 1,$$

we can maximize the SNR by setting $w(x,y)=I'(x,y)$  So, the optimal weight function is the (unknown) profile of the object itself.

Since we don't have that, what we want to use for a weight function is something as similar as possible to the profile of the object we want to measure.

A good choice is a Gaussian that has the same moments as the function we're fitting.
