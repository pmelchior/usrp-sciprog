# How Telescopes Work

## The Basics

We'll focus on optical/UV/IR telescopes - these all work mostly the same way:
1. Photons hit one or more mirrors (the first mirror - the largest - is called the *primary*).
2. Photons are redirected slightly by (much smaller) lenses.
3a. For *imaging*: photons with wavelengths we don't want are absorbed or reflected by a *filter*.
3b. For *spectroscopy*: Photons go through a slit mask or fiber to block out those that aren't from an object we care about, and then are redirected according to their wavelength by a diffraction grating or prism.
4. Photons hit a semiconductor detector (typically *CCDs* - Charged Coupled Devices) and convert into electrons.  Those electrons are held in place (within their *pixels*) by many tiny electric fields.
5. After waiting a bit (the *exposure time*) we close a shutter, blocking the light, and *read out* the detector by converting the number of electrons in each pixel into a digital signal.

   - Pro: Infrared (helps to be in space)
   - Pro: both spectra and imaging
   - Pro: Very good angular resolution (0.1")
   - Pro: Many fewer time-of-day constraints
   - Con: Impossible to maintain
   - Con: Small FoV
   - Con: No UV or optical
   - Con: \$10 billion
   - Pro/Con: complicated (dangerous) launch and deployment
 - http://en.wikipedia.org/wiki/Giant_Magellan_Telescope
   - Pro: Good angular resolution (with adaptive optics)
   - Pro: Large wavelength coverage
   - Pro: both spectra and imaging
   - Pro: it's HUGE - easier to go deeper
   - Pro: multiple instruments, easier maintenance
   - Con: atmosphere in the way
   - Con: not a huge field of view
   - Con: time-of-day, weather constraints
   - Pro/Con: ~\$1 billion
 - http://en.wikipedia.org/wiki/Large_Synoptic_Survey_Telescope
   - Pro: huge etendue: FoV * mirror size (big survey)
   - Con: pretty good resolution (0.7")
   - Pro: good for transients (short exposure time)
   - Con: smaller collecting area
   - Pro/Con: ~\$1 billion
   - Con: expensive to process the data
   - Con: no adaptive optics
   - Con: just one instrument (imaging, no spectra)
