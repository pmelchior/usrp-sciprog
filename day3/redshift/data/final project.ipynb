```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.coordinates as coord
import astropy.units as u
from astropy import constants as const
```

```python
data = np.genfromtxt('obj4.data', names=True)

print(data)
```

```python
redshifts = data['Z']

velocities = redshifts * 3e5

print(redshifts)

print(velocities)

meanVel = np.mean(velocities)

print(meanVel)

plt.rcParams.update({'font.size': 20,
                     'xtick.labelsize' : 18,
                     'ytick.labelsize' : 18})

plt.figure(figsize=(10,10))
plt.hist(redshifts)
plt.xlabel('z')
plt.ylabel('number of data points')
plt.title('Histogram of Redshifts in Dataset')

plt.figure(figsize=(10,10))
plt.hist(velocities)
plt.xlabel('velocity (km/s)')
plt.ylabel('number of data points')
plt.title('Histogram of Velocities in Dataset')
```

```python
xCoords = data['RAJ2000']
yCoords = data['DECJ2000']

# print(data.dtype.names)

plt.figure(figsize=(15,15))
plt.scatter(xCoords, yCoords, c= redshifts)
plt.colorbar()
plt.xlabel('RA (deg)')
plt.ylabel('DEC (deg)')
plt.title('Locations of all Data Points Colored by Redshift')
```

```python
plt.figure(figsize=(15,15))
plt.scatter(xCoords, yCoords, c= velocities-meanVel)
plt.colorbar()

plt.xlabel('RA (deg)')
plt.ylabel('DEC (deg)')
plt.title('Locations of all Data Points Colored by Velocity')
```

```python
cutData = data[(data['Z'] < 0.002)]

meanRedshiftCutData = np.mean(cutData['Z'])
print(meanRedshiftCutData)

# print(data.dtype.names)

plt.figure(figsize=(15,15))
plt.scatter(cutData['RAJ2000'], cutData['DECJ2000'], c= (cutData['Z'] - meanRedshiftCutData) * 3e5)
plt.colorbar()
plt.xlabel('RA (deg)')
plt.ylabel('DEC (deg)')
plt.title('Velocity Dispersion for Low-Redshift Points')

plt.figure(figsize=(10,10))
plt.hist((cutData['Z'] - meanRedshiftCutData) * 3e5)
plt.xlabel('velocity dispersion (km/s)')
plt.ylabel('number of data points')
plt.title('Velocity Dispersion of Low-Redshift Points')

print(np.mean((abs(cutData['Z'] - meanRedshiftCutData)) * 3e5))

plt.figure(figsize=(10,10))

plt.hist((cutData['Z']) * 3e5)
plt.xlabel('velocity (km/s)')
plt.ylabel('number of data points')
plt.title('Absolute Velocities of Low-Redshift Points')
```
